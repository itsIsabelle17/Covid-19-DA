### TIME DIMENSION (Date ID Foreign key for 3 Tables)

* Create Date Table

```python
%%sql
DROP TABLE IF EXISTS date CASCADE;

CREATE TABLE date (
    Date_ID SERIAL PRIMARY KEY,
    hour CHAR(19),
    day CHAR(10),
    year INTEGER,
    month_of_year_str VARCHAR(12),
    month_of_year INTEGER,
    day_of_month INTEGER,
    day_of_week_str CHAR(9),
    day_of_week INTEGER,
    hour_of_day INTEGER,
    quarter_of_year INTEGER
);
```

```python
%%sql
INSERT INTO Date (hour, day, year, month_of_year_str, month_of_year, day_of_month, 
                  day_of_week_str, day_of_week, 
                  hour_of_day, quarter_of_year)
SELECT DISTINCT TO_CHAR(last_update, 'YYYY-MM-DD HH24:00:00') AS hour,
    TO_CHAR(last_update, 'YYYY-MM-DD') AS day, 
    CAST(TO_CHAR(last_update, 'YYYY') AS INTEGER) AS year,
    TO_CHAR(last_update, 'Month') AS month_of_year_str,
    CAST(TO_CHAR(last_update, 'MM') AS INTEGER) AS month_of_year,
    CAST(TO_CHAR(last_update, 'DD') AS INTEGER) AS day_of_month,
    TO_CHAR(last_update, 'Day') AS day_of_week_str,
    CAST(TO_CHAR(last_update, 'D') AS INTEGER) AS day_of_week,
    CAST(TO_CHAR(last_update, 'HH24') AS INTEGER) AS hour_of_day,
    CAST(TO_CHAR(last_update, 'Q') AS INTEGER) AS quarter_of_year
FROM cases
union
SELECT DISTINCT TO_CHAR(date, 'YYYY-MM-DD HH24:00:00') AS hour,
    TO_CHAR(date, 'YYYY-MM-DD') AS day, 
    CAST(TO_CHAR(date, 'YYYY') AS INTEGER) AS year,
    TO_CHAR(date, 'Month') AS month_of_year_str,
    CAST(TO_CHAR(date, 'MM') AS INTEGER) AS month_of_year,
    CAST(TO_CHAR(date, 'DD') AS INTEGER) AS day_of_month,
    TO_CHAR(date, 'Day') AS day_of_week_str,
    CAST(TO_CHAR(date, 'D') AS INTEGER) AS day_of_week,
    CAST(TO_CHAR(date, 'HH24') AS INTEGER) AS hour_of_day,
    CAST(TO_CHAR(date, 'Q') AS INTEGER) AS quarter_of_year
FROM vaccine
union
SELECT DISTINCT TO_CHAR(date, 'YYYY-MM-DD HH24:00:00') AS hour,
    TO_CHAR(date, 'YYYY-MM-DD') AS day, 
    CAST(TO_CHAR(date, 'YYYY') AS INTEGER) AS year,
    TO_CHAR(date, 'Month') AS month_of_year_str,
    CAST(TO_CHAR(date, 'MM') AS INTEGER) AS month_of_year,
    CAST(TO_CHAR(date, 'DD') AS INTEGER) AS day_of_month,
    TO_CHAR(date, 'Day') AS day_of_week_str,
    CAST(TO_CHAR(date, 'D') AS INTEGER) AS day_of_week,
    CAST(TO_CHAR(date, 'HH24') AS INTEGER) AS hour_of_day,
    CAST(TO_CHAR(date, 'Q') AS INTEGER) AS quarter_of_year
FROM times
ORDER by day;
```

#### Q1. Whats the percentage of people who are affected by COVID based on state ?

```python
%%sql
SELECT cases.fips1, state.province_state, state.population,
SUM(round(cases.incident_rate,2)) as incident_rate, round((SUM(cases.incident_rate) / cast(state.population as decimal(10,2))*100),4) as percentage
FROM state,cases
WHERE state.fips = cases.fips1 and cases.date_id = 689
GROUP BY cases.fips1, state.province_state, state.population
ORDER BY percentage limit 10 ;
```

* From this data we can infer that california has the highest percentage of incident_rate.

```python
%%sql
SELECT cases.fips1, state.province_state, state.population,
SUM(cases.confirmed) as total_cases_confirmed, round((SUM(cases.confirmed) / cast(state.population as decimal(10,2))*100),4) as percentage 
FROM state,cases
WHERE state.fips = cases.fips1 and cases.date_id = 689 and state.fips != '60' and state.fips != '78' and state.fips != '66' and state.fips != '69'
GROUP BY cases.fips1, state.province_state, state.population
ORDER BY percentage DESC limit 10;
```

* As we can see from the above table North Dakota and Alaska have higher covid affected people since they arestates with lower population under 1 million people. All top 10 states with highest covid percentage of covid confirmed cases besides Tennessee are all under 5 million popoluation.

```python
%%sql
SELECT cases.state_id,state.province_state, max(confirmed) Max_Confirmed, max(deaths) Max_Death, max(recovered1)Max_Recovered
FROM cases 
INNER JOIN state ON cases.state_id = state.state_id
WHERE fips1 < 57
GROUP BY cases.state_id, state.province_state
ORDER BY province_state ASC
LIMIT 5;
```

```python
%%sql
SELECT date, vaccine.province_state, fully_vaccinated, partially_vaccinated, state.population,
round((SUM(vaccine.fully_vaccinated) / cast(state.population as decimal(10,2))*100),2) as percentage
FROM vaccine
INNER JOIN state on vaccine.state_id=state.state_id
WHERE date between '2021-11-01' and '2021-12-01' AND (vaccine.fips = 36) AND vaccine_type = 'All'
GROUP BY date, vaccine.province_state, state.population, vaccine.partially_vaccinated, vaccine.fully_vaccinated
ORDER BY percentage limit 10 ;
```

```python
%%sql
SELECT date, vaccine.province_state, fully_vaccinated, partially_vaccinated, state.population,
round((SUM(vaccine.fully_vaccinated) / cast(state.population as decimal(10,2))*100),2) as percentage
FROM vaccine
INNER JOIN state on vaccine.state_id=state.state_id
WHERE date between '2021-11-01' and '2021-12-01' AND (vaccine.fips = 48) AND vaccine_type = 'All'
GROUP BY date, vaccine.province_state, state.population, vaccine.partially_vaccinated, vaccine.fully_vaccinated
ORDER BY percentage limit 10 ;
```

This is a positive effect of vaccinations having an impact of lowering the case fatality rate.
